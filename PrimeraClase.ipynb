### Ordenamiento Burbuja

```python
numeros = [8,10,6,12,2,4]

intercambio = True
while intercambio:
    intercambio = False
    for i in range(len(numeros) - 1):
        if numeros[i] > numeros[i + 1]:
            intercambio = True
            numeros[i], numeros[i + 1] = numeros[i + 1], numeros[i]

print(numeros)
```

```python
#Ordenamiento por Selección
numeros = [8,10,6,2,12,4]

for i in range(len(numeros)):
  for j in range(i+1,len(numeros)):
    if numeros[i] > numeros[j]:
      numeros[i],numeros[j]=numeros[j],numeros[i]

print(numeros)
```

### El Método Definitivo: .sort() vs sorted()

#### sort()

```python
#Es un método. Ordena la lista original (in-place).

notas = [45, 89, 32, 100, 75, 89]
notas.sort(reverse=True)
print("La nota más alta fue: ",notas[0])
print("El top 3 es: ",notas[:3])
```

#### sorted()

```python
#Es una función. Te devuelve una copia ordenada.

notas = [45, 89, 32, 100, 75, 89]

notasRanking = sorted(notas, reverse=True)

print("La nota más alta fue:", notasRanking[0])
print("El Top 3 es: ", notasRanking[:3])
```

### La funcion zip()

```python
paises = ["México", "España", "Colombia"]
capitales = ["CDMX", "Madrid", "Bogotá"]

# zip une las listas, dict las convierte en diccionario
mapa = dict(zip(paises, capitales))
print(mapa)
```

### Métodos y funciones de los diccionarios

```python
#El metodo keys()
for key in mapa.keys():
    print(key, "->", mapa[key])
```

```python
#El metodo values()
for valor in mapa.values():
  print(valor)
```

```python
#El metodo items(), lo convierte en una lista de tuplas
for llave, valor in mapa.items():
  print(llave," -> ",valor)
```

### Modificar, agregar y eliminar valores

```python
#Agregando nuevas claves
mapa['Bolivia']='La Paz'
print(mapa)

#El método .update(), Si se quiere agregar varios elementos a la vez
mapa.update({'Brasil':'Brasilia','Chile':'San Tiago','Argentina':'Buenos Aires'})
print(mapa)
```

```python
#El método .pop() (Eliminar y extraer)
eliminado = mapa.pop('Colombia')
print(eliminado)
print(type(eliminado))
print(mapa)
```

```python
#El método .popitem() (Eliminar el último)
llave,valor = mapa.popitem()
print('Se elimino: ',llave,' -> ',valor)
print(mapa)
```

```python
#El método .clear() (Vaciar todo)
mapa.clear()
print(mapa)
```
